### I/O: writing data to and from a file

``` python
from sys import argv

script, arg1 = argv

def read():
    fh = open(raw_input("Filename to read:"), 'r')
    for line in fh:
        print line
    fh.close()

def write():
    new_name = raw_input("Name of a file to write:")
    fnew = open(new_name, 'w')
    new_name.write("Hello world! My code in Python gets more and more advanced!")
    fnew.close()
    
if arg=='read':
    read()
elif arg=='write':
    write()
else:
    print "USAGE: python script.py keyword \n where keyword is read or write."


```

```python
#write_data.py
#we will save data to a text file and a binary file
import numpy as np
from sys import argv
script, arg1 = argv
database = np.array((['Greta', 60], ['John', 50])) #Name, Age

def save(sth):
    np.savetxt('./database.txt', database)
    print "Saved a file."

def load():
    data = np.genfromtxt('./database.txt', names=['name', 'age'])
    print "Loaded a file. \n"
    print "Names in the file: ", data['name']
    print "Ages in the file: ", data['age']


    
if arg=='save':
    save()
elif arg=='load':
    load()
else:
    print "USAGE: python write_data.py keyword \n where keyword is save or load."


```
